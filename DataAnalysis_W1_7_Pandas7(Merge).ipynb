```python
# Importing libraries
import pandas as pd
```

```python
# Loading datasets
# 'olympic_2022' va 'olympic_2024' nomlariga datasetlarni yuklab oling
olympic_2022 = pd.read_csv('./OlympicGames2022.csv')
olympic_2024 = pd.read_csv('./OlympicGames2024.csv')
```

```python
olympic_2022
```

```python
olympic_2024
```

```python
# Data Analysis Part
# 2022 yilgi ma'lumotlar bazasiga 'NOC' nomli ustun yarating va davlatlarning qisqartirilgan shaklini yozing

olympic_2022['NOC'] = ['USA', 'CHA']
```

```python
# 2024 yilgi ma'lumotlar bazasiga ham 'NOC' nomli ustun yarating va davlatlarning qisqartirilgan shaklini yozing
```

```python
# Ikkita datasetni 'merge' metodi yordamida birlashtiring
pd.merge(olympic_2022, olympic_2024)
```

```python
# 'on' argumentini 'Country' ga qo'llab jadval hosil qiling
pd.merge(olympic_2022, olympic_2024, on='Country')
```

```python
# 'suffixes' dan foydalanib ustun nomlarini o'zgartiring
pd.merge(olympic_2022, olympic_2024, on='Country', suffixes=['_2022', '_2024'])
```

```python
# 'how' argumnetidan foydalanib 'inner'uchun db hosil qiling
pd.merge(olympic_2022, olympic_2024, on='Country', suffixes=['_2022', '_2024'], how='inner')
```

```python
# 'how' argumnetidan foydalanib 'left'uchun db hosil qiling
pd.merge(olympic_2022, olympic_2024, on='Country', suffixes=['_2022', '_2024'], how='left')
```

```python
# 'how' argumnetidan foydalanib 'right'uchun db hosil qiling
pd.merge(olympic_2022, olympic_2024, on='Country', suffixes=['_2022', '_2024'], how='right')
```

```python
# 'how' argumnetidan foydalanib 'outer'uchun db hosil qiling
result = pd.merge(olympic_2022, olympic_2024, on='Country', suffixes=['_2022', '_2024'], how='outer')
result
```

```python
# NaN qiymatlani 0 bilan almashtirib chiqing
result.fillna(0, inplace=True)
result
```

```python
# 'Total' nomli ustun yaratib unga jami 2022 va 2024 dagi medallarni hisoblab yozing
result['Total_Gold'] = result['Gold_2022'] + result['Gold_2024']
result['Total_Silver'] = result['Silver_2022'] + result['Silver_2024']
result
```

```python
# 'Rank' nomli ustun yaratib, shu yillar mobaynida eng ko'p 'oltin'
# medal olgan davlatlarga o'rin bering
result = result.sort_values('Total_Gold')
result
```
