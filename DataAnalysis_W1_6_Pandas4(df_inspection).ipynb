## Tavsiflash (description) va Xulosa chiqarish
### Xulosa statistikalarini olish


````python
df.describe()
````

5<sup>th</sup> va 95<sup>th</sup> percentillarini belgilash:


````python
df.describe(percentiles=[0.05, 0.95])
````

````python
df.describe(include=np.object)
````

Barcha ma'lumot turlari uchun

````python
df.describe(include='all')
````

Yuqoridagi amal ustunlar uchun ham ishlaydi

````python
df.felt.describe()
````

Maxsus statistikalar uchun metodlar ham mavjud. Quyida ularning ba'zilari:

| Metod | Tavsif | Ma'lumot turlari |
| --- | --- | --- |
| `count()` | Null bo'lmagan kuzatuvlar soni | Har qanday |
| `nunique()` | Yagona qiymatlar soni | Har qanday |
| `sum()` | Qiymatlar yig'indisi | Raqamli yoki Boolean |
| `mean()` | Qiymatlarning o'rtacha qiymati | Raqamli yoki Boolean |
| `median()` | Qiymatlarning medianasi | Raqamli |
| `min()` | Qiymatlarning eng kichigi | Raqamli |
| `idxmin()` | Eng kichik qiymatning indeksini topish | Raqamli |
| `max()` | Qiymatlarning eng kattasi | Raqamli |
| `idxmax()` | Eng katta qiymatning indeksini topish | Raqamli |
| `abs()` | Ma'lumotlarning mutlaq qiymatlari | Raqamli |
| `std()` | Standart og'ish | Raqamli |
| `var()` | Variansiya |  Raqamli |
| `cov()` | Ikki `Series` orasidagi kovariatsiya yoki `DataFrame`da barcha ustun kombinatsiyalari uchun kovariatsiya matritsasi | Raqamli |
| `corr()` | Ikki `Series` orasidagi korrelyatsiya yoki `DataFrame`da barcha ustun kombinatsiyalari uchun korrelyatsiya matritsasi | Raqamli |
| `quantile()` | Maxsus kvantilni hisoblash | Raqamli |
| `cumsum()` | Kumulativ yig'indi | Raqamli yoki Boolean |
| `cummin()` | Kumulativ eng kichik qiymat | Raqamli |
| `cummax()` | Kumulativ eng katta qiymat | Raqamli |
