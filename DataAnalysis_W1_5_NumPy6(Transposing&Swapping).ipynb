## Massivlar pozitsiyasini ko'chirish `Transpose` va o'qlar o'rnini almashtirish `Swapping axes`

```python
A = np.arange(6).reshape(3, -1) # A massiv elementlari 0-6 gacha, shape (3, 2)
A
```

## `Transpose`

```python
A.T # A massivning Transpose ko'rinishini beradi
```

## `Swapping`

```python
A
```

```python
A.swapaxes(1, 0) # A massivning Transpose ko'rinishini beradi
```
